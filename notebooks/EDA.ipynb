# EXPLORATORY DATA ANALYSIS 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the data
file_path = r"C:\Users\user\loan-default-analytics\Default_Fin.csv"
df = pd.read_csv(file_path)

# 2. Drop the unnecessary Index column
df = df.drop(columns=['Index'])

# 3. Check for missing values
print("Missing Values per Column:")
print(df.isnull().sum())
print("\n" + "="*40 + "\n")

# 4. Check class distribution of our target variable
print("Loan Default Distribution:")
print(df['Defaulted?'].value_counts(normalize=True) * 100)
```

## Visualize the Risk Factors


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for clean visualizations
sns.set_theme(style="whitegrid")

# Create a figure with two subplots side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. Boxplot for Bank Balance vs Defaulted
sns.boxplot(x='Defaulted?', y='Bank Balance', data=df, ax=axes[0], palette="Set2")
axes[0].set_title('Bank Balance by Default Status')
axes[0].set_xticklabels(['No Default (0)', 'Default (1)'])

# 2. Boxplot for Annual Salary vs Defaulted
sns.boxplot(x='Defaulted?', y='Annual Salary', data=df, ax=axes[1], palette="Set2")
axes[1].set_title('Annual Salary by Default Status')
axes[1].set_xticklabels(['No Default (0)', 'Default (1)'])

plt.tight_layout()
plt.show()
```

### Key Insights
1. Bank Balance is a Massive Risk FactorLook at the plot on the left. The difference is night and day!Customers who do not default (0) have a much lower median bank balance (around $10,000$).Customers who default (1) have a significantly higher bank balance (median well over $20,000$).
This suggests that customers carrying higher outstanding monthly bank balances are under much greater financial stress and are far more likely to default on their loans. This will be a heavily weighted feature for our Machine Learning model!
2. Annual Salary Tells a Surprising Story
Look at the plot on the right. The two boxes are almost entirely aligned. The median salary for people who default is nearly identical to those who don't.
Salary alone doesn't actually tell us if someone will default. A high earner who mismanages their balance is just as risky as a low earner.
